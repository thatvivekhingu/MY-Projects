```python
real_ratings = ratings[ratings['User-ID'].isin(y)]
```

```python
real_ratings.head(2)
```

```python
ratings_on_books = real_ratings.merge(books,on='Book-Id')
```

```python
ratings_on_books.head(2)
```

```python
number_rating = ratings_on_books.groupby('Book-name')['Ratings'].count().reset_index()
```

```python
real_ratings.head(1)
```

```python
#umber_rating = number_rating.drop(columns=['Ratings'])
final_ratings = ratings_on_books.merge(number_rating,on='Book-name')
```

```python
final_ratings.head(2)
```

```python
final_ratings = final_ratings.rename(columns={'num_of_Ratings': 'Ratings'})

```

```python
final_ratings.drop(columns=['Ratings_y'],inplace=True)
```

```python
final_ratings.head()
```

```python
final_ratings.rename(columns={'Ratings_x':'Ratings'
,'Ratings':'num_of_ratings'},inplace=True)
```

```python
final_ratings.sample(7)
```

```python
final_ratings.drop_duplicates(['User-ID','Book-name'],inplace=True)
```

```python
final_ratings.shape
```

```python
book_pivot=final_ratings.pivot_table(columns='User-ID',index='Book-name',values='Ratings')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot.head(3)
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestion =model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
  print(book_pivot.index[suggestion[i]])
```

```python
import os
import pickle

# Create the artifacts directory if it doesn't exist
if not os.path.exists('artifacts'):
    os.makedirs('artifacts')

pickle.dump(model,open('artifacts/model.pkl','wb'))
pickle.dump(book_name,open('artifacts/book_name.pkl','wb'))
pickle.dump(final_ratings,open('artifacts/final_ratings.pkl','wb'))
pickle.dump(book_pivot,open('artifacts/book_pivot.pkl','wb'))
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distance,suggestion =model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

  for i in range(len(suggestion)):
    books=book_pivot.index[suggestion[i]]
    for j in books:
      print(j)
```

```python
book_name=='A Whisper of Blood'
recommend_book(book_name)
```
